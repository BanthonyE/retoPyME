```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('tweet_emotions.csv')
df.columns = ['id', 'tag', 'text']
df.sample(5)
```

```
df.drop('id', axis=1, inplace=True)
```

```
df.shape
```

```
df['tag'].value_counts()
```

```
df.tag.unique().shape
```

```
df['text'] = df['text'].apply(lambda x: x.lower())
```

```
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if not word.startswith(('@', 'http'))]))
```

```
remove = [',', '.', '-', '\'', '!', '?', '¿', '¡', '&', '#', '$', '%', '=', '½']
for i in remove:
    df['text'] = df['text'].apply(lambda x: x.replace(i, ''))
```

```
df.sample(5)
```

```
df.tag.replace(['sadness', 'worry', 'hate', 'boredom', 'anger'], 'negative', inplace=True)
```

```
df.tag.replace(['happiness', 'love', 'fun', 'relief', 'enthusiasm'], 'positive', inplace=True)
```

```
df.tag.replace(['empty', 'surprise'], 'neutral', inplace=True)
```

```
df.tag.value_counts()
```

```
df_balanced = pd.concat([df[df.tag == 'positive'].sample(11000),
              df[df.tag == 'negative'].sample(11000),
              df[df.tag == 'neutral'].sample(11000)], axis=0)
```

```
df_balanced.sample(3)
```

```
df_balanced.tag.value_counts()
```

```
df_balanced.to_csv('balanced.csv', index=False)
```
